```python
%matplotlib notebook
%matplotlib inline
import blimpy
import glob
from astropy.time import Time
import matplotlib.pyplot as pp
import sys

# glob.glob takes a path with wildcards (like *) and expands the path into a list of paths matched by the wildcard
# the following glob produces a list of ~500 paths
files=sorted(glob.glob("/mnt_blpd11/datax/moonbounce_splices/*/spliced*.0002.fil")) # find files
files=files[:10]
print("List of files produced by glob")
print(len(files))
print(type(files))
print(files)
print("end of list")
badfiles=["/mnt_blpd11/datax/moonbounce_splices/AGBT19B_033_03/spliced_guppi_58765_00861_MOON_0001.gpuspec.0000.fil",
          "/mnt_blpd11/datax/moonbounce_splices/AGBT19B_033_03/spliced_guppi_58765_01174_MOON_0002.gpuspec.0000.fil"]
#for file in files[:5]: #getslice with first 5 files
pp.close()
pp.figure(figsize=(6.4,len(files)*3.5))
filenum=1
for file in files: #getslice with first 5 files
    print(filenum, file) # print filename
    p=file.split("/")
    obs=p[-2]
    t=p[-1][14:].split(".")[0]
    pp.subplot(len(files),1,filenum)
    title=(obs + "-" + t)
    w = blimpy.Waterfall(filename=file)
    w.plot_spectrum()
    filenum+=1
    pp.ylim(0,5e10)
pp.savefig("x.pdf")
pp.savefig("x.png")
pp.tight_layout()
pp.show()
```
